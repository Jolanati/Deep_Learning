## Section 2: Tensors — The Universal Container for Data

---

### What Is a Tensor?

Before an AI model can learn anything, all data — whether it's text, sales figures, images, or audio — must be converted into numbers arranged in a structured grid. That grid is called a **tensor**.

Think of a tensor as an upgraded spreadsheet:

| Structure | Everyday equivalent | Business example |
|-----------|-------------------|-----------------|
| A single row of numbers | One row in a spreadsheet | Daily revenue: 120, 340, 89 |
| A table (rows and columns) | A standard spreadsheet | Monthly sales by product and region |
| A stack of tables | Multiple spreadsheet tabs | 2 years of monthly sales data |
| A 4-level structure | A spreadsheet with multiple views | A batch of 8 product photos, each in colour |

The deeper the structure, the more information it holds — but the logic is the same: organised numbers in a grid.

---

### Why Not Just Use a Spreadsheet?

Spreadsheets are great for human analysis, but they break down at scale. An AI model might process:

- 60,000 training images
- Each image made up of 784 individual pixel values
- Across thousands of training cycles

A spreadsheet cannot handle that. Tensors are designed specifically to work at that scale — and to run on a GPU, which can process millions of numbers simultaneously.

---

### Four Ways to Create a Tensor

When building an AI system, data is loaded into tensors in different ways depending on where it comes from:

- **Random numbers** — used to initialise a model before training begins. The model starts with random guesses and gradually improves. Like assigning placeholder values to a new employee's scorecard before any real data comes in.

- **All zeros** — used to create blank templates or reset values. Like starting a fresh budget spreadsheet with no entries yet.

- **All ones** — used in certain calculations as a neutral starting point.

- **Your own data** — when you have real numbers (prices, pixel values, sensor readings), you load them directly into a tensor, just as you would paste data into a spreadsheet.

---

### What to Expect When This Runs

The code below creates four different tensors and prints them. You will see grids of numbers on screen. The exact values of the random tensor will be different each time you run it — that is intentional and expected.


```python
import torch

# A 2-row, 3-column table of random numbers between 0 and 1
tensor_random = torch.rand((2, 3))

# A 4x4 table of zeros — like an empty template
tensor_zeros = torch.zeros((4, 4))

# A 3x3 table of ones
tensor_ones = torch.ones((3, 3))

# You supply the exact numbers
tensor_list = torch.tensor([[1, 2, 3], [4, 5, 6]])

print("Random (2x3):\n", tensor_random)
print("\nZeros (4x4):\n", tensor_zeros)
print("\nOnes (3x3):\n", tensor_ones)
print("\nFrom list (2x3):\n", tensor_list)
```

## Section 3: Doing Maths on Tensors — Three Operations You Will See Constantly

---

### Why Maths on Tensors Matters

Every time an AI model makes a prediction, it is doing maths on tensors. Understanding the *type* of maths being done helps you understand what the model is actually doing under the hood.

There are three operations you will encounter repeatedly:

---

### Operation 1: Adding Two Grids Together

Imagine you have two spreadsheets with the same structure — say, January costs and January revenues for the same set of products. Adding them together means adding each cell to its matching cell in the other sheet.

That is exactly what this operation does. Every number in the first grid is added to the number in the exact same position in the second grid. The result is a new grid of the same size.

> **Business use:** Combining two data sources that cover the same dimensions — for example, merging predicted values with actual values to calculate error at each position.

---

### Operation 2: Multiplying Two Grids Together (Position by Position)

Same logic as addition, but with multiplication. Each cell in the first grid is multiplied by the cell in the same position in the second grid.

> **Business use:** Applying a grid of weights or scaling factors to a grid of values — like multiplying each product's sales volume by its individual profit margin to get profit per product.

---

### Operation 3: Matrix Multiplication — The Core of AI

This one is different, and it is the most important operation in all of deep learning.

Instead of pairing up matching cells, matrix multiplication combines each **row** of the first grid with each **column** of the second grid — multiplying them together and summing the results.

> **Business analogy:** Think of a weighted scoring model. You have a table of candidates (rows) each scored on several criteria, and a table of criteria weights (columns). Matrix multiplication gives you each candidate's final weighted score in one operation — across all candidates and all weights simultaneously.

This is how neural networks apply learned weights to input data to produce predictions. Every layer of a neural network is essentially a matrix multiplication.

---

### What to Expect When This Runs

The code below performs all three operations and prints the results. Pay attention to the last section which highlights the difference between simple position-by-position multiplication (result: 5) and matrix multiplication (result: 19) — same inputs, completely different outcomes.


```python
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

# Each position added to its matching position
c = a + b
print("Element-wise addition (a + b):\n", c)

# Each position multiplied by its matching position
d = a * b
print("\nElement-wise multiplication (a * b):\n", d)

# Matrix multiplication: rows × columns (linear algebra)
e = torch.matmul(a, b)
print("\nMatrix multiplication torch.matmul(a, b):\n", e)

print("\n--- Key difference ---")
print("a * b gives:", d[0,0].item(), "(1×5 = 5)")
print("matmul gives:", e[0,0].item(), "(1×5 + 2×7 = 19)")
```

## Section 4: Inspecting a Tensor — The Three Questions You Always Ask

---

### Why You Need to Inspect Your Data

In any data project, before you process a dataset, you ask basic questions: How big is it? What format is it in? Where is it stored? Getting these wrong causes errors — or worse, silently incorrect results.

Working with tensors is no different. There are three properties you check constantly, especially when something is not working as expected.

---

### The Three Properties

**Size and Structure**

The first thing to check is the shape — how many rows, columns, and layers does this tensor have? A mismatch in shape is one of the most common causes of errors. For example, trying to combine a dataset with 1,000 rows with a set of weights designed for 500 rows will fail immediately.

> Like checking that two spreadsheets have the same number of columns before trying to merge them.

**Data Type**

Tensors can store different kinds of numbers — whole numbers (integers) or decimal numbers (floating point). Certain AI operations require decimal numbers. If you feed in whole numbers by mistake, the calculation may fail or produce incorrect results.

> Like the difference between a column formatted as "Text" vs "Number" in Excel — it looks the same to the eye, but the formula behaves differently.

**Location**

Data can live in standard memory (CPU) or on the GPU. Operations can only be performed between tensors that are in the same location. Trying to combine a tensor on the CPU with one on the GPU will cause an error.

> Like trying to run a formula that references a file on a server that is currently offline — the data exists, but it is not accessible from where you are working.

---

### What to Expect When This Runs

The code below creates a tensor and prints its three key properties. This is a habit every data scientist builds — always inspect before you process.


```python
tensor = torch.rand(2, 3)

print("Tensor:\n", tensor)
print("\n--- Properties ---")
print("Shape  :", tensor.shape)   # How big is this grid?
print("Dtype  :", tensor.dtype)   # What type of numbers?
print("Device :", tensor.device)  # Where is it stored?
```

In a production AI environment — for example, a company running real-time fraud detection or image classification at scale — the difference between CPU and GPU processing can be the difference between results in 2 hours vs results in 3 days.

## Section 6: Reshaping Data — Same Numbers, Different Structure

---

### Why Would You Need to Reshape Data?

AI models are strict about the shape of data they receive. A model designed to process images expects data in a very specific structure. Raw data rarely arrives in exactly the right format — so reshaping is one of the most frequently performed operations in any AI pipeline.

The important thing to understand is: **reshaping does not change any values**. It only reorganises how they are arranged.

---

### A Business Analogy

Imagine you have 24 monthly revenue figures for 2 years. Depending on what you want to do with them, you might arrange them differently:

- **As a single list:** All 24 numbers in one row — useful for sending to a calculation engine
- **As a table:** 4 rows × 6 columns — useful for a quarterly view
- **As a 3D structure:** 2 pages × 3 rows × 4 columns — useful for comparing Year 1 vs Year 2 by quarter

In all three cases, the 24 numbers are identical. Only the organisation changes.

This is exactly what reshaping does with tensors — and it is essential because different layers of a neural network expect different structures.

---

### The "Figure It Out Yourself" Shortcut

One particularly useful feature: if you know all dimensions except one, you can tell the system to calculate the missing dimension automatically. Rather than doing the maths yourself, you simply leave a placeholder and the system fills it in. This prevents calculation errors and makes the code more robust when data sizes change.

---

### What to Expect When This Runs

The code takes 24 numbers and demonstrates three different ways to organise them. You will see the same data presented as a flat list, a 2D table, and a 3D structure. The values never change — only the shape.


```python
t = torch.arange(24)
print("Original — single list, shape:", t.shape)
print(t)

t_2d = t.reshape(4, 6)
print("\nAs a 4×6 table:", t_2d.shape)
print(t_2d)

t_flat = t_2d.reshape(-1)
print("\nFlattened back to list:", t_flat.shape)

t_3d = t.reshape(2, 3, 4)
print("\nAs a 3D structure (2 pages × 3 rows × 4 columns):", t_3d.shape)
print(t_3d)
```

## Section 7: Fine-Tuning Structure — Adding, Removing, and Reordering Dimensions

---

### Three Small but Critical Adjustments

Beyond basic reshaping, there are three operations used constantly to prepare data for specific model layers. They sound technical but the logic behind each one is very practical.

---

### Removing Unnecessary Dimensions

Sometimes data ends up with a redundant layer of structure — a dimension that contains only one item. This happens often when data is processed through a pipeline that expects a certain format.

> **Analogy:** Imagine a folder on your desktop that contains only one other folder, which then contains the actual file. The outer folder serves no purpose — you remove it to get directly to what you need.

This operation removes those "wrapper" dimensions automatically.

---

### Adding a Batch Dimension

AI models are designed to process data in batches — groups of items at once, rather than one at a time. This is more efficient. However, if you want to run a single image through a model for a quick test, you need to tell the model: *"This is a batch of one."*

Adding a batch dimension does exactly that — it wraps a single item in the structure of a batch, so the model accepts it without modification.

> **Analogy:** A courier service designed to handle pallets of boxes. If you only have one box, you still need to put it on a pallet before they will process it.

---

### Reordering Dimensions

Different tools and libraries expect data in different orders. PyTorch stores images as: **Batch → Colour channels → Height → Width**. Some other tools expect: **Batch → Height → Width → Colour channels**. The data is identical — only the order of axes changes.

> **Analogy:** Transposing a spreadsheet. All the same data, just rows and columns swapped to match what the next tool expects.

---

### What to Expect When This Runs

The code demonstrates all three operations on small example tensors. Focus on the shape printed before and after each step — the numbers inside do not change, only how they are organised.


```python
t = torch.rand(3, 1, 4)
print("Original shape:", t.shape)  # has a redundant dimension of size 1

t_sq = t.squeeze()
print("After removing redundant dimension:", t_sq.shape)  # cleaner structure

t_unsq = t_sq.unsqueeze(0)
print("After adding batch dimension:", t_unsq.shape)  # ready for a model

img = torch.rand(8, 3, 32, 32)  # 8 images, channel-first format
img_hwc = img.permute(0, 2, 3, 1)  # reorder to channel-last format
print("\nOriginal image batch shape (channel-first):", img.shape)
print("Reordered image batch shape (channel-last):", img_hwc.shape)
```

## Section 8: Building a Custom Dataset — Telling the System Where Your Data Lives

---

### The Challenge at Scale

In a real business AI project, your data does not arrive pre-packaged. You have a folder of images, a database of records, or a collection of documents — and you need to connect that raw source to the AI training pipeline.

PyTorch handles this through a concept called a **Dataset**. A Dataset is essentially a formal description of your data: it tells the system how many items exist and how to retrieve any individual item on demand.

---

### Why "On Demand" Matters

A naive approach would be to load all your data into memory at the start. For small datasets this works. For large ones — say 500,000 product images — this would immediately exhaust your computer's memory and crash the process.

The Dataset approach solves this by loading only what is needed, when it is needed. The training process asks for item number 4,217 — the Dataset fetches exactly that item and returns it. No more, no less.

> **Business analogy:** Think of a warehouse management system. Instead of moving every product to the factory floor at the start of the day, the system retrieves each item from the shelf exactly when it is needed on the production line. Efficient, scalable, and nothing is wasted.

---

### What a Dataset Must Be Able to Answer

Every custom Dataset must answer two questions:

1. **"How many items do you have?"** — So the training loop knows when it has seen all the data once (one full pass through the data is called an "epoch").

2. **"Give me item number X."** — So the training loop can retrieve any specific sample, apply any preprocessing, and pass it to the model.

That is it. Those two capabilities are the entire contract between your data and the AI system.

---

### What to Expect When This Runs

The code below defines a Dataset class for a folder of images. It does not run any training — it simply establishes the connection between a folder on disk and the AI pipeline. You will not see any output unless you connect it to a real folder, which is covered in the next section.


```python
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os

class CustomImageDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.image_files = os.listdir(image_dir)
        self.transform = transform

    def __len__(self):
        # Answer to: "how many items do you have?"
        return len(self.image_files)

    def __getitem__(self, index):
        # Answer to: "give me item number X"
        img_path = os.path.join(self.image_dir, self.image_files[index])
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image

print("CustomImageDataset class defined — ready to connect to a folder of images.")
```

## Section 9: The DataLoader — Delivering Data Efficiently to the Model

---

### The Problem It Solves

Once a Dataset knows where your data is and how to retrieve items, you still need to decide *how* that data is delivered to the model during training. Do you send one item at a time? All at once? In random order?

The **DataLoader** handles all of this automatically. It sits between your Dataset and your training loop, acting as a logistics layer.

---

### Three Key Decisions the DataLoader Makes

**Batch size — how many items to process at once**

Models do not learn from one example at a time — they learn from groups. Processing 32 images at once is more computationally efficient than processing them one by one, and it also leads to more stable learning.

> **Analogy:** A factory production line processes items in batches, not one unit at a time. The batch size is how many units go through each cycle.

**Shuffle — whether to randomise the order each time**

If the model always sees the same data in the same order, it can start to memorise the sequence rather than learn the underlying patterns. Randomising the order each time the dataset is looped through prevents this.

> **Analogy:** A teacher who always calls on students in alphabetical order — the students learn to anticipate their turn rather than genuinely learning the material. Randomising keeps everyone engaged and learning from real understanding.

**Number of workers — how many processes load data in parallel**

While the model is processing one batch, the next batch can be loaded in the background by a separate process. This prevents the GPU from sitting idle while it waits for data.

> **Analogy:** A restaurant kitchen where one team cooks while another team preps the next order in parallel — no downtime between dishes.

## Section 10: Loading a Real Dataset — MNIST as a Starting Point

---

### What Is MNIST?

MNIST is one of the most well-known benchmark datasets in AI. It contains 70,000 handwritten digit images — numbers 0 through 9, each in a small 28×28 pixel format.

- **60,000 images** are used for training — the model learns from these
- **10,000 images** are used for testing — the model is evaluated on these, which it has never seen before

> **Business analogy:** Think of MNIST as the "hello world" of image recognition — the standard starting point that every AI practitioner uses to verify that their pipeline is working correctly before moving to more complex, proprietary data.

---

### The Transform Pipeline — Preparing Data Before It Enters the Model

Raw images cannot go directly into a neural network. They need to be standardised first. This is done through a **transform pipeline** — a sequence of steps applied automatically to every image as it is loaded.

Two essential transforms are applied here:

**Converting to a number grid**
The raw image (a picture file) is converted into a grid of decimal numbers between 0 and 1, where 0 = black and 1 = white. This is necessary because the model only understands numbers, not image files.

**Normalising the range**
The values are then shifted from the range [0 to 1] to the range [-1 to 1]. This centring of values around zero helps the model learn more quickly and stably — a well-established practice backed by decades of machine learning research.

> **Analogy:** Before running a statistical analysis, a data analyst might normalise all variables to the same scale so that one large-valued column does not dominate the results. Same principle here.

---

### What to Expect When This Runs

- MNIST will be downloaded automatically the first time (a small file, under 12 MB)
- The output confirms 60,000 training samples and shows the shape of a single batch: 32 images, each 1 colour channel (greyscale), 28 pixels tall, 28 pixels wide


```python
import torchvision.transforms as transforms
from torchvision import datasets
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.ToTensor(),          # image file → numbers in [0, 1]
    transforms.Normalize(0.5, 0.5)  # shift to [-1, 1] for stable training
])

train_dataset = datasets.MNIST(root='data', train=True, transform=transform, download=True)
test_dataset  = datasets.MNIST(root='data', train=False, transform=transform, download=True)
train_loader  = DataLoader(train_dataset, batch_size=32, shuffle=True)

print(f"Training samples : {len(train_dataset):,}")
print(f"Test samples     : {len(test_dataset):,}")
print(f"Batches per epoch: {len(train_loader)}")

images, labels = next(iter(train_loader))
print(f"\nOne batch — images: {images.shape}, labels: {labels.shape}")
print("Image shape means: (32 images, 1 colour channel, 28px height, 28px width)")
```

## Section 11: Data Augmentation — Making Your Data Work Harder

---

### The Business Problem This Solves

Training an AI model requires a large and varied dataset. In an ideal world, you would collect thousands of examples covering every possible real-world scenario — different lighting conditions, angles, sizes, and backgrounds.

In practice, collecting that much data is expensive and time-consuming. **Data augmentation** is the solution: it artificially increases the diversity of your training data by applying random variations to the images you already have.

The model sees a slightly different version of each image every time it trains — which forces it to learn the *underlying pattern* rather than memorising specific examples.

---

### Why This Matters for Business

Without augmentation, an AI model trained on clean studio photos of products might fail when it encounters photos taken in poor lighting or at an unusual angle — common in real-world usage.

With augmentation, the model is exposed to simulated versions of those variations during training, making it far more robust in production.

> **Analogy:** Training a new employee on only perfect, textbook scenarios leaves them unprepared for real-world situations. Exposing them to varied, sometimes messy cases during training produces a more resilient and capable employee.

---

### What Augmentations Are Applied

| Transformation | What it does | Why it helps |
|----------------|--------------|--------------|
| Resize | Standardises all images to the same dimensions | Models require consistent input sizes |
| Random horizontal flip | Mirrors the image left-right 50% of the time | The model learns that orientation does not change the identity of an object |
| Random rotation | Tilts the image slightly | Builds robustness to objects that are not perfectly upright |
| Colour jitter | Slightly varies brightness and contrast | Simulates different lighting conditions |
| Normalise | Scales pixel values to a consistent range | Stabilises and accelerates learning |

**Important:** Augmentation is applied only to training data — never to test or validation data. When evaluating the model, you want to see its performance on clean, unmodified images, not artificially altered ones.

---

### What to Expect When This Runs

No output is produced — this cell simply defines the two transform pipelines (training and validation). The next sections will use them.


```python
import torchvision.transforms as transforms

# Training pipeline — augmentation applied to increase variety
train_transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(degrees=15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# Validation pipeline — no augmentation, clean images only
val_transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

print("Transform pipelines defined.")
print("Training pipeline : 6 steps including augmentation")
print("Validation pipeline: 3 steps, no augmentation")
```

## Section 12: Visualising a Batch — Seeing What the Model Actually Receives

### Why Visualisation Matters

Before displaying, the process is reversed: each pixel value is scaled back to the displayable range. This is purely for human viewing — the model itself always works with the normalised values.

> **Analogy:** A financial report prepared in thousands (000s) for internal use needs to be converted back to full figures before presenting to an external audience. The underlying data is the same — just the presentation scale changes.

### What to Expect When This Runs

A 2-row grid of 16 handwritten digit images will appear. Above each image is the correct label (0–9). You will notice that some images appear slightly tilted or flipped — that is the augmentation working as intended. Every time you run this cell, you will see a different random batch.


```python
import matplotlib.pyplot as plt
from torchvision import datasets
from torch.utils.data import DataLoader

augmented_dataset = datasets.MNIST(root='data', train=True, transform=train_transform, download=False)
loader_augmented  = DataLoader(augmented_dataset, batch_size=32, shuffle=True)

imgs, labels = next(iter(loader_augmented))

fig, axes = plt.subplots(2, 8, figsize=(14, 4))
fig.suptitle("Sample training batch — augmented MNIST (16 of 32 shown)", fontsize=11)

for i, ax in enumerate(axes.flat):
    img = imgs[i].squeeze().numpy()
    img = (img * 0.5) + 0.5  # reverse normalisation for display
    ax.imshow(img, cmap='gray', vmin=0, vmax=1)
    ax.set_title(str(labels[i].item()), fontsize=9)
    ax.axis('off')

plt.tight_layout()
plt.show()
print("Each title number = the correct label for that image.")
print("Notice: some images are slightly rotated or flipped — augmentation at work.")
```
