## Section 1: Setting Up — Importing PyTorch and Checking What Hardware You Have

Imagine you're about to run a massive data analysis — millions of rows, thousands of calculations per second. Before you open the file, you need to:
1. Check that your software is installed
2. Check whether you have a powerful machine to run it on

That's exactly what this first cell does.

---

**`import torch`** — loads the PyTorch library into memory. Without this line, nothing else works. It's like opening the application before you can use it.

**`import torchvision`** — loads a companion package that gives you ready-made datasets (like MNIST) and image transformation tools. Think of it as the plugin pack for image work.

**`torch.cuda.is_available()`** — this is the important one. It asks: *"Does this machine have a GPU?"*

- A **CPU** (your regular processor) is like a highly skilled consultant — great at complex, sequential tasks, but you only have a few of them.
- A **GPU** (a graphics chip, originally built for video games) is like an army of junior analysts — thousands of them, each doing a simple task simultaneously. For deep learning, this is 10 to 100 times faster.

If a GPU is available, the code will use it automatically. If not, it falls back to the CPU — and your code still runs, just slower.

> Think of it like a delivery company automatically routing a large shipment to their biggest truck if one is available, or a smaller van if not — same destination, different speed.


```python
import torch
import torchvision

print("PyTorch version  :", torch.__version__)
print("Torchvision version:", torchvision.__version__)
print("GPU available    :", torch.cuda.is_available())
print("Device to use    :", "cuda" if torch.cuda.is_available() else "cpu")
```

## Section 2: Creating Tensors

**What is a tensor?**

A tensor is just a container for numbers, organised in a grid. The grid can have any number of dimensions:

| Shape | What it looks like | Real-world example |
|-------|-------------------|-------------------|
| `(3,)` | A single row: `[1, 2, 3]` | One day of sales figures |
| `(3, 4)` | A table: 3 rows, 4 columns | A small spreadsheet |
| `(2, 3, 4)` | A stack of tables | 2 months of weekly sales |
| `(8, 3, 32, 32)` | A batch of colour images | 8 photos, each 32×32 pixels, 3 colour channels |

**The four ways to create a tensor:**

- `torch.rand((2, 3))` — fills a 2×3 grid with random numbers between 0 and 1. Like randomly filling a table with percentages.
- `torch.zeros((4, 4))` — fills a 4×4 grid with all zeros. Like an empty template.
- `torch.ones((3, 3))` — fills a 3×3 grid with all ones.
- `torch.tensor([[1,2,3],[4,5,6]])` — you provide the exact numbers. Like typing values directly into a spreadsheet.


```python
import torch

# A 2-row, 3-column table of random numbers between 0 and 1
tensor_random = torch.rand((2, 3))

# A 4x4 table of zeros — like an empty template
tensor_zeros = torch.zeros((4, 4))

# A 3x3 table of ones
tensor_ones = torch.ones((3, 3))

# You supply the exact numbers
tensor_list = torch.tensor([[1, 2, 3], [4, 5, 6]])

print("Random (2x3):\n", tensor_random)
print("\nZeros (4x4):\n", tensor_zeros)
print("\nOnes (3x3):\n", tensor_ones)
print("\nFrom list (2x3):\n", tensor_list)
```

## Section 3: Basic Tensor Operations

**Three types of maths on tensors:**

**1. Element-wise addition (`a + b`)**
Adds matching positions. Position (0,0) of `a` + position (0,0) of `b`, and so on.
> Like adding two spreadsheets cell by cell: Jan revenue + Jan costs = Jan result.

**2. Element-wise multiplication (`a * b`)**
Multiplies matching positions. Same idea — each number multiplies its counterpart.
> Like applying a percentage adjustment to each cell individually.

**3. Matrix multiplication (`torch.matmul(a, b)`)**
This is the classic linear algebra operation — rows of `a` multiplied by columns of `b` and summed up. This is NOT the same as element-wise multiplication.
> Like a weighted scoring model: each row of criteria × each column of weights = final scores.

The key difference: `a * b` is position-by-position. `torch.matmul(a, b)` combines rows with columns.


```python
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

# Each position added to its matching position
c = a + b
print("Element-wise addition (a + b):\n", c)

# Each position multiplied by its matching position
d = a * b
print("\nElement-wise multiplication (a * b):\n", d)

# Matrix multiplication: rows × columns (linear algebra)
e = torch.matmul(a, b)
print("\nMatrix multiplication torch.matmul(a, b):\n", e)

print("\n--- Key difference ---")
print("a * b gives:", d[0,0].item(), "(1×5 = 5)")
print("matmul gives:", e[0,0].item(), "(1×5 + 2×7 = 19)")
```

## Section 4: Inspecting Tensor Properties

When something goes wrong in deep learning, the first thing you check is the shape, type, and location of your data. These three attributes are your debugging toolkit:

| Attribute | What it tells you | Example |
|-----------|------------------|---------|
| `.shape` | How many rows, columns, dimensions | `torch.Size([2, 3])` = 2 rows, 3 columns |
| `.dtype` | What kind of numbers are stored | `torch.float32` = decimal numbers |
| `.device` | Where the data lives | `cpu` or `cuda:0` (GPU) |

> **Business analogy:** It's like checking a report's format before processing it — how many rows does it have? Are the numbers integers or decimals? Is it on the local server or in the cloud?

**Why does `.dtype` matter?** Some operations require all tensors to have the same type. Mixing `float32` and `int64` can cause errors — just like mixing currencies in a financial model.


```python
tensor = torch.rand(2, 3)

print("Tensor:\n", tensor)
print("\n--- Properties ---")
print("Shape  :", tensor.shape)   # How big is this grid?
print("Dtype  :", tensor.dtype)   # What type of numbers?
print("Device :", tensor.device)  # Where is it stored?
```
